```python
class Stack:
    def __init__(self):
        self.head=None
        self.tail=None
        self.count = 0

    def is_empty(self):
        return self.head is None
         
    def push(self,data):
        new_node=Node(data)
        if self.is_empty():
            self.head=new_node
            self.tail=new_node
        else:
            new_node.next=self.head
            self.head=new_node
        self.count+=1
        print(f"{data},was pushed into the stack")
        
    def pop(self):
        if self.is_empty():
            print("Error")
            return None
        self.count-=1
        node_to_delete = self.head
        if self.head==self.tail:
            self.head=None
            self.tail=None
        else:
            self.head = self.head.next

        print(f"popped item is {node_to_delete.data}")    
        
    def peek(self):
        if self.is_empty():
            print("Error")
            return None
        print(f"top of the stack is {self.head}")
```

```python
class queue:
    def __init__(self):
        self.head=None
        self.tail=None
        self.count=0
    def is_empty(self):
        return self.head is None
        
    def enqueue(self,data):
        new_node=Node(data)
        if self.is_empty():
            self.head=new_node
            self.tail=new_node

        else:
            new_node.next=self.head
            self.head=new_node
        print(f"enqueued item is {data}") 
        
    def dequeue(self):
        if self.is_empty():
            print("error:stack is empty")
            return None
        if self.head==self.tail:
            node_data=self.head.data
            self.head=None
            self.tail=None
            return f"dequeud item is {node_data}"

        prev = self.head
        while prev.next !=self.tail:
            prev=prev.next
        node_data = self.tail.data
        prev.next=None
        self.tail=prev
        return f"dequeued item is {node_data} "
        
```
