```python
class operand_stack:
    stack = []
    def __init__(self,expression):
        self.expression = expression

    def check(self):
        tokens = self.expression.split()
        for token in tokens:
            if token.isdigit():
                num=int(token)
                stack[].append(num)
            else:
                stack.append(token)
                

```

```python
expn = '3 4 + 2 *'
k = operand_stack(expn)
k.check()
```
